```python
# notebooks/weather_eda.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Load the data
df = pd.read_csv('../data/processed/live_weather.csv')

# Convert datetime column to datetime type
df['datetime'] = pd.to_datetime(df['datetime'])

# Set style
sns.set(style="whitegrid")

# Show first few rows
df.head()
```

```python
# Check for missing values
df.isnull().sum()

# Basic statistics
df.describe()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='datetime', y='temperature', data=df, color='red')
plt.title('Temperature Over Time')
plt.xlabel('Time')
plt.ylabel('Temperature (Â°C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Humidity
plt.figure(figsize=(12, 4))
sns.lineplot(x='datetime', y='humidity', data=df)
plt.title('Humidity Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Pressure
plt.figure(figsize=(12, 4))
sns.lineplot(x='datetime', y='pressure', data=df)
plt.title('Pressure Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count plot of weather types
plt.figure(figsize=(10, 4))
sns.countplot(x='weather', data=df)
plt.title('Weather Conditions Frequency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
